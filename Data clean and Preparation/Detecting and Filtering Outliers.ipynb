```python
#检测和过滤异常值；
import numpy as np
import pandas as pd

data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
#查询第三列中绝对值大于3的值；
col = data[2]
col[np.abs(col) > 3]
```

```python
#选择满足这样条件的行：行中任意有一个元素的绝对值大于3，使用any()方法；
data[(np.abs(data) > 3).any(1)]
```

```python
#cap values outside the interval–3 to 3，覆盖-3和3之外的值；
#numpy.sign,The sign function returns -1 if x < 0, 0 if x==0, 1 if x > 0. nan is returned for nan inputs;
data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
```
